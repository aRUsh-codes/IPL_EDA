## **Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
deliveries=pd.read_csv("/content/deliveries.csv")
```

```python
matches=pd.read_csv("/content/matches.csv")
```

```python
merged=pd.merge(deliveries,matches,left_on="match_id",right_on="id")
```

## **Deliveries Data Preprocessing**

```python
deliveries.head(10)
```

```python
deliveries.info()
```

```python
deliveries.describe()
```

```python
print("Count and percentage")
l=deliveries.isnull().sum()
m=deliveries.isnull().sum()*100/len(deliveries)
nll=pd.concat([l,m],axis=1,keys=("Count","Percentage"))
nll
```

```python
deliveries.drop(["dismissal_kind","fielder"],axis=1,inplace=True)
```

## **Matches Data Preprocessing**

```python
matches.head()
```

```python
matches.info()
```

```python
print("Count and percentage")
l1=matches.isnull().sum()
m1=matches.isnull().sum()*100/len(matches)
nll1=pd.concat([l1,m1],axis=1,keys=("Count","Percentage"))
nll1
```

```python
matches.drop(["umpire3"],axis=1,inplace=True)
```

```python
factors=["winner","player_of_match","venue","umpire1"]
for factor in factors:
  print(factor.upper(),"\n",pd.DataFrame(matches[factor].value_counts()[:5]))
  print("\n")
```

```python
print("The data starts from year {} to {}".format(matches["season"].min(),matches["season"].max()))
```

## **Exploratory Data Analysis**

```python
plt.figure(figsize=(10,10))
ax=sns.barplot(x=matches["winner"].value_counts()[:10].values,y=matches["winner"].value_counts()[:10].index,palette="Spectral")
plt.show()
```

*Mumbai Indians has won most IPL matches.*

```python
plt.figure(figsize=(10,10))
ax=sns.barplot(x=matches["player_of_match"].value_counts()[:10].values,y=matches["player_of_match"].value_counts()[:10].index,palette="RdBu")
plt.show()
```

*CH Gayle has been awarded the **player of the match** the most times.*

```python
matches["toss_decision"].nunique()
```

```python
right_decision=[]
for i in range(len(matches["winner"])):
  if matches["toss_winner"][i]==matches["winner"][i]:
    right_decision.append(1)
  else:
    right_decision.append(0)
matches["right_decision"]=right_decision
matches.head()
```

*Dropping "id","city","date,"umpire1" and "umpire2"*

```python
i=["id","city","date","umpire1","umpire2"]
for j in i:
  matches.drop([j],axis=1,inplace=True)
```

```python
plt.figure(figsize=(10,7))
ax=sns.countplot(x="toss_decision",hue="right_decision",data=matches,color="orchid")
for p in ax.patches:
  ax.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
              ha = 'center', va = 'center', xytext = (3, 10), textcoords = 'offset points')
plt.show()
```
